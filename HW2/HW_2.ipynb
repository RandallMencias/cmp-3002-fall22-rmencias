# Time Decorator (HW1)

```python
import time
import random
def clock(func):
    def inner(*arg):
        start = time.time() * 10 ** 3
        answer = func(*arg)
        time.sleep(0.1)
        end = time.time() * 10 ** 3
        #print("Execution Time:" + str(end - start))
        return answer, (end - start)

    return inner
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import ctypes
class Stack(object):
    """
    Implementation of the stack data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.stack = self._create_stack(self.n)

    def _create_stack(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()

    def push(self, item):
        """
            Add new item to the stack
            """
        if self.l == self.n:
            raise ValueError("no more capacity")
        self.stack[self.l] = item
        self.l += 1

    def pop(self):
        """
            Remove an element from the stack
            """
        # self.l = 0
        # 0 is equivalent to False
        # any number != 0 is True
        if not self.l:
            raise 'stack is empty'
        c = self.stack[self.l - 1]
        # self.stack[self.l] = ctypes.py_object
        self.l -= 1
        return c

    def top(self):
        """
            Show the top element of the stack
            """
        return self.stack[self.l - 1]

    def full(self):
        """
            Is the stack full?
            """
        return self.l == self.n
        # if self.l == self.n:
        #    return True
        # return False

    def empty(self):
        """
            Is the stack empty?
            """
        return self.l == 0
        # if self.l == 0:
        #    return True
        # return False

    def size(self):
        """
            Return size of the stack
            """
        return self.l

    def print(self):
        for i in range(self.l):
            print(self.stack[i])
```

```python
class Singly_linked_list(Singly_linked_list):
    def reverse(self):
        """
         Reverses list
                """
        node = self.head_node
        prev = None # Creates variable for previuos node, O(1) of extra memory
        while node is not None: # original head needs to end on None
            next = node.next_node #creates a next variable
            node.next_node = prev # prev node.next_node is assigned as next node
            prev = node #inverst position between node and next node
            node = next #moves on to the next set of nodes
        self.head_node = prev # head becomes last prev node


S = Singly_linked_list()
for i in range(1,5):
    S.insert_head(Node(i))
S.list_traversed()
S.reverse()
print("Reversed:")
S.list_traversed()
```

### 3. Implement the class Queue using stacks.


```python

class Queue(object):
    def __init__(self, n):
        self.n = n
        self.stack = self._create_queue(self.n)

    def _create_queue(self, n): #creates queue using a Stack
        return Stack(n)

    def enqueue(self, item):
        """
        Adds item to queue
        Input: item
                        """
        self.stack.push(item)
    @clock
    def deque(self):
        """
        Returns head and removes it
        Output: Stack head
                        """
        temp = Stack(self.stack.l) #creates a temp stack
        head = self.stack.top() #gets top value
        for i in range(self.stack.l-1):
            temp.push(self.stack.top())
            self.stack.pop() #reverses the stack value into temp except for the last value
        self.stack.pop() #empties stack
        for i in range(temp.l):
            self.stack.push(temp.top()) #reverses values back into self stack
            temp.pop() #empties temp
        return head



    def first(self):
        return self.stack.stack[0]

    def full(self):
        return self.stack.l == self.stack.n

    def empty(self):
        return self.stack.l == 0

    def size(self):
        return self.stack.l

    def print(self):
        for i in range(0, self.stack.l):
            print(self.stack.stack[i])



S = Queue(5)
S.enqueue(1)
S.enqueue(2)
S.enqueue(4)
S.enqueue(-1)
S.print()
print("Deque:")
S.deque()
S.print()
```
